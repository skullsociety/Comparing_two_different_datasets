## Use Instructions

Function:  
1. compare_df_columns(oldmnr, newmnr)
2. common_column_difference_total_number(oldmnr,newmnr)
3. column_totalrows_variableschange_newretiredvariables(oldmnr,newmnr,variable_count_limit= None)
4. column_variable_frequency(oldmnr,newmnr,variable_count_limit= None)
5. all(oldmnr,newmnr,variable_count_limit= None)

Function 1: compare_df_columns(oldmnr,newmnr)  
Compares new version's columns against old version's Columns.  
Output would have Retired columns, New Columns, Common columns.  

| Output         | Old DF | New DF|
|--------------|:-----:|-----------:|
| Retired Columns      |  Y    |    N       |
| New Columns     |  N |          Y |
| Common Columns      |  Y |          Y |


Function 2: common_column_difference_total_number(oldmnr,newmnr)  
Comapres each column total rows, output would be in statement whether there is a difference between the total rows.  
Serves as a Summary

Function 3: column_totalrows_variableschange_newretiredvariables(oldmnr,newmnr,variable_count_limit= None)
Compares each columns, new and old df  
Variable_count_limit would define how much variable would it be required before the column would not be considered  
to run by the code. Able to use this to cut down on the number of columns to run, cutting down on time and computational resources. Default value for variable_count_limit is 20.  

exmaple output for each column:  
[cdmp_diagnoses_list] column  
__________________________________________________  
New MNR [cdmp_diagnoses_list] column have 0 change in total rows/values.  
There is no change in column variables in cdmp_diagnoses_list  
Too many unique column variables (>1000)  
Code is cut short to save computing time and resources  


Function 4: column_variable_frequency(oldmnr,newmnr,variable_count_limit= None)  
Compares the variables of a column and the frequency of each variables in each column.  
Default value for variable_count_limit is 20. If output empty DF, there is no differences.  
Otherwise, it would output "Different Column value variable, New_MNR_Frequency, Old_MNR_Frequency, Frequency difference".  

Function 5: all(oldmnr,newmnr,variable_count_limit= None)  
Function 1 - 4

### Comparing Difference of columns between MNR

```python
#defination that compares columns.
#output retired columns in old mnr and new columns in new mnr
#retired columns are columns in old mnr but not in new mnr
#new columns are columns in new mnr but not in old mnr
def compare_df_columns(oldmnr, newmnr):
    #putting old MNR column into a lists
    oldmnrcolumns= []
    for i in oldmnr.columns:
        oldmnrcolumns.append(i)

    #old mnr columns
    oldmnr.columns


    #putting old MNR columns into a lists
    newmnrcolumns= []
    for i in newmnr.columns:
        newmnrcolumns.append(i)


    #new mnr columns
    newmnr.columns


    #columns in old mnr, but not in new mnr
    retired_columns = []
    #columns in new mnr but not in old mnr
    new_columns = []
    #common columns 
    common_columns=[]

    #compare between lists
    for column in oldmnrcolumns:
        if column not in newmnrcolumns:
            retired_columns.append(column)

    for column in newmnrcolumns:
        if column not in oldmnrcolumns:
            new_columns.append(column)
            
    for column in newmnrcolumns:
        if column in oldmnrcolumns:
            common_columns.append(column)

    #sort lists
    retired_columns.sort()
    new_columns.sort()
    common_columns.sort()

    print(f'''Retired Columns: {retired_columns}
    
New Columns: {new_columns}

Common Columns: {common_columns}
    ''')
```

```python
def column_totalrows_variableschange_newretiredvariables(oldmnr,newmnr,variable_count_limit= None):
    
    #Old mnr columns list
    oldmnrcolumns= []
    for i in oldmnr.columns:
        oldmnrcolumns.append(i)

    #new mnr columns list
    newmnrcolumns= []
    for i in newmnr.columns:
        newmnrcolumns.append(i)

    #common columns list
    common_columns=[]

    for column in newmnrcolumns:
        if column in oldmnrcolumns:
            common_columns.append(column)

    common_columns.sort()
    
    #check for number of variables in column and compare what variables are missing
    selectiveprint = ''
    
    for i in range(len(common_columns)):
        numberofvariables_totalcount_change = (newmnr[common_columns[i]].nunique() - oldmnr[common_columns[i]].nunique())
        if numberofvariables_totalcount_change == 0:
            selectiveprint =  f'There is no change in column variables in {common_columns[i]}'
        elif numberofvariables_totalcount_change < 0:
            selectiveprint = f'There is {numberofvariables_totalcount_change*-1} less column variables in new mnr'
        elif numberofvariables_totalcount_change > 0:
            selectiveprint = f'There is {numberofvariables_totalcount_change} more column variables in new mnr'

    





        #if number of variable in columns is less than variable_count_limit, find out what is the variable added or removed
        if variable_count_limit is None:
            variable_count_limit = 20

        oldmnruniquevariable = []
        newmnruniquevariable = []
        newuniquevariable= []
        retireduniquevariable = []
        commonuniquevariable = []

        if numberofvariables_totalcount_change <= variable_count_limit:
            #Putting all new and old columns variable into lists
            for x in newmnr[common_columns[i]].unique():
                #condition if number of unique values is more than 1000, break out of loop, for identifier columns like resident_id
                if len(newmnruniquevariable) >1000:
                    break
                else:
                    newmnruniquevariable.append(x)
            for x in oldmnr[common_columns[i]].unique():
                if len(oldmnruniquevariable) >1000:
                    break
                else:
                    oldmnruniquevariable.append(x)

            #filter new, common and retired unique column variables into lists list
            #Sort out the unique variables only if unique variables is less than 1000
            if len(newmnruniquevariable) < 1000 and len(oldmnruniquevariable) <1000:
                for x in newmnruniquevariable:
                    if x not in oldmnruniquevariable:
                        newuniquevariable.append(x)
                    elif x in oldmnruniquevariable:
                        commonuniquevariable.append(x)
                    elif x not in newuniquevariable:
                        retireduniquevariable.append(x)

            #Generation of report, dependent on the number of difference in mnr variables
            selectiveprint1 = ''
            selectiveprint2 = ''
            selectiveprint3 = ''

            if len(newmnruniquevariable) < 1000 and len(oldmnruniquevariable) <1000:

                if len(newuniquevariable) > 0:
                    selectiveprint1 = f'''
Number of new unique variables: {len(newuniquevariable)},
New unique variables: {newuniquevariable}
        '''
                else:
                     selectiveprint1 = f'Number of new unique variables: {len(newuniquevariable)}'


                if len(retireduniquevariable) > 0:
                    selectiveprint2 = f'''
Number of retired variables: {len(retireduniquevariable)},
retired variables: {retireduniquevariable}
        '''
                else:
                     selectiveprint2 = f'Number of retired variables: {len(retireduniquevariable)}'

                if len(commonuniquevariable) <= variable_count_limit:
                    selectiveprint3 = f'''
Number of common variables: {len(commonuniquevariable)},
common variables: {commonuniquevariable}
    '''
                else:
                    selectiveprint3 = f'More than {variable_count_limit} common variables'

            else:
                selectiveprint1 = f'Too many unique column variables (>1000)'
                selectiveprint2 = f'Code is cut short to save computing time and resources'
                selectiveprint3 = f' '

            #report
            print(f'''
[{common_columns[i]}] column
__________________________________________________
New MNR [{common_columns[i]}] column have {numberofvariables_totalcount_change} change in total rows/values.
{selectiveprint}
{selectiveprint1}
{selectiveprint2}
{selectiveprint3}
    ''')


        else:
            print(f'There are more than {variable_count_limit} in differences of number of variable in new mnr in [{common_columns[i]}] column')


```

## Compare column variable frequency

```python
def column_variable_frequency(oldmnr,newmnr,variable_count_limit= None):
    
    #Old mnr columns list
    oldmnrcolumns= []
    for i in oldmnr.columns:
        oldmnrcolumns.append(i)

    #new mnr columns list
    newmnrcolumns= []
    for i in newmnr.columns:
        newmnrcolumns.append(i)

    #common columns list
    common_columns=[]

    for column in newmnrcolumns:
        if column in oldmnrcolumns:
            common_columns.append(column)

    common_columns.sort()
    
    
    
    
    
    if variable_count_limit is None:
            variable_count_limit = 20
    

    

    

    data = {'Column value variable':[0], 'Frequency':[0]}
    new_mnr_column_frequency = pd.DataFrame(columns=['Column value variable','Frequency'])
    old_mnr_column_frequency = pd.DataFrame(columns=['Column value variable','Frequency'])
    
    for i in range(len(common_columns)):
        selectiveprint4= ''
        selectiveprint5= ''
        print(f'\n[{common_columns[i]}] column\n__________________________________________________\n')
              
              
        new_mnr_sorted_frequency_field_list = newmnr[common_columns[i]].value_counts().index.tolist()
        
        newmnrsortedfrequencylistcount = []
        oldmnrsortedfrequencylistcount = []
        
        #Condition to abort counting if it is too big
        if len(new_mnr_sorted_frequency_field_list) <= variable_count_limit:
            #new mnr columns sorted frequency without index from high frequency first
            for x in range(len(new_mnr_sorted_frequency_field_list)):
                newmnrsortedfrequencylistcount.append(newmnr[common_columns[i]].value_counts()[x])
                #create a dataframe with index and frequency/count
            data = {'Column value variable':new_mnr_sorted_frequency_field_list, 'Frequency':newmnrsortedfrequencylistcount}
            new_mnr_column_frequency =pd.DataFrame(data)

        else:
            print(f'More than {variable_count_limit} variable limit in new mnr column {common_columns[i]}')
        
        
        #old mnr
        #old mnr columns index list, sorted by higher frequency first
        old_mnr_sorted_frequency_field_list = oldmnr[common_columns[i]].value_counts().index.tolist()

        #Condition to abort counting if it is too big
        if len(old_mnr_sorted_frequency_field_list) <= variable_count_limit:
            #new mnr columns sorted frequency without index from high frequency first
            for x in range(len(old_mnr_sorted_frequency_field_list)):
                oldmnrsortedfrequencylistcount.append(oldmnr[common_columns[i]].value_counts()[x])
                #create a dataframe with index and frequency/count
            data = {'Column value variable':old_mnr_sorted_frequency_field_list, 'Frequency':oldmnrsortedfrequencylistcount}
            old_mnr_column_frequency =pd.DataFrame(data)

        else:
            print(f'More than {variable_count_limit} variable limit in old mnr column {common_columns[i]}')
            

            
            
            
        #compare the variable frequency between new and old. Formula = New mnr Frequency - old MNR frequency
        column_variable_different_frequency_fields = []
        column_variable_different_frequency_number = []
        New_column_variable_different_frequency_number = []
        Old_column_variable_different_frequency_number = []
       
        
        for x in range(len(new_mnr_column_frequency['Frequency'])):
            column_variable_different_frequency_number = new_mnr_column_frequency['Frequency'][x] - old_mnr_column_frequency['Frequency'][x]
            if column_variable_different_frequency_number != 0:
                column_variable_different_frequency_fields.append(new_mnr_column_frequency['Column value variable'][x])
                column_variable_different_frequency_number.append(column_variable_different_frequency_number)
                New_column_variable_different_frequency_number.append(new_mnr_column_frequency['Frequency'][x])
                Old_column_variable_different_frequency_number.append(old_mnr_column_frequency['Frequency'][x])

        #put difference between column variables and it's frequencies in a dataframe
        data = {'Different Column value variable':column_variable_different_frequency_fields, 'New_MNR_Frequency': New_column_variable_different_frequency_number, 'Old_MNR_Frequency' :Old_column_variable_different_frequency_number, 'Frequency difference':column_variable_different_frequency_number}
        difference_mnr_column_frequency =pd.DataFrame(data)
        print(difference_mnr_column_frequency)

        #Crafting report variable for difference in variable frequency
        selectivetext5=''
        if difference_mnr_column_frequency.empty:
            selectivetext5='There is no difference between variables frequency'
            print(selectivetext5)
        else:
            selectivetext5 = difference_mnr_column_frequency
            print(selectivetext5)
```

## Run all

```python
def all (oldmnr, newmnr,variable_count_limit= None):
    compare_df_columns(oldmnr, newmnr)
    common_column_difference_total_number(oldmnr,newmnr)
    column_totalrows_variableschange_newretiredvariables(oldmnr,newmnr,variable_count_limit= variable_count_limit)
    column_variable_frequency(oldmnr,newmnr,variable_count_limit= variable_count_limit)
```
